# 🤖 Cómo funciona

Esta plantilla lanza tu **primer agente de IA**, un chatbot impulsado por inteligencia artificial que puede hacer más que solo hablar:  
**puede tomar acción usando herramientas.**

Piensa en un agente de IA como un **asistente inteligente**, y las herramientas como las **aplicaciones en tu teléfono**.  
Al conectarlo con otros nodos, le permites interactuar con datos y servicios del mundo real, como **consultar el tiempo**, **obtener noticias** o incluso **enviar correos electrónicos** en tu nombre.

---

## 🧩 Estructura del flujo de trabajo

Este flujo de trabajo está diseñado para ser el punto de partida perfecto:

### 💬 La interfaz de chat
Un **Chat Trigger node** proporciona una interfaz sencilla y limpia para que puedas hablar con tu agente.

### 🧠 El cerebro
El **AI Agent node** recibe tus mensajes, decide inteligentemente qué herramienta usar (si es necesario) y formula una respuesta útil.  
Su personalidad e instrucciones son totalmente personalizables en el campo **"Mensaje del Sistema"**.

### 🗣️ El modelo de lenguaje
Utiliza **Google Gemini** para potenciar sus habilidades de razonamiento y conversación.

### 🧰 Las herramientas
Viene preequipado con dos herramientas para demostrar sus capacidades:

- 🌦️ **Obtener el tiempo** → obtiene previsiones meteorológicas en tiempo real.  
- 📰 **Obtener noticias** → lee cualquier fuente RSS para obtener los últimos titulares.

### 🧠 La memoria
Un **Conversation Memory node** permite al agente recordar los últimos mensajes, lo que habilita **conversaciones naturales y de seguimiento**.

## ⚙️ Pasos de configuración

⏱️ **Tiempo estimado:** ~2 minutos  

Solo necesitas una cosa para comenzar:  
una **clave API gratuita de Google AI**.

---

### 🔑 Obtén tu clave API de Google AI

1. Visita [Google AI Studio](https://aistudio.google.com/app/apikey).  
2. Haz clic en **"Crear clave API en nuevo proyecto"** y copia la clave que aparece.

---

### 🔧 Agrega tus credenciales en n8n

1. En el lienzo del flujo de trabajo, ve al nodo **Connect your model (Google Gemini)**.  
2. Haz clic en el menú desplegable **Credential** y selecciona **+ Create New Credential**.  
3. Pega tu clave API en el campo **Clave API** y haz clic en **Guardar**.

---

## 💬 ¡Empieza a chatear!

1. Ve al nodo **Example Chat**.  
2. Haz clic en el botón **"Abrir chat"** en su panel de parámetros.  
3. Intenta hacerle una de las preguntas de ejemplo, como:
   - “¿Qué tiempo hace en París?”
   - “Consígueme las últimas noticias tecnológicas.”

---

## 🚀 ¡Listo!
Ya tienes un **agente de IA completamente funcional**.  
Prueba a añadir más herramientas (como **Gmail** o **Google Calendar**) para que sea aún más potente.

🧠 Qué es un RSS Feed

Un RSS feed es un enlace que entrega las últimas noticias en formato XML.
El nodo “Get news” de tu agente ya está preparado para leer ese tipo de fuentes (como las de BBC, CNN o TechCrunch).
Solo tenemos que darle la URL RSS de Infobae.
